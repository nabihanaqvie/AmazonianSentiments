```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
fig, axs = plt.subplots(ncols=5)
j  = sns.regplot(x=BlendDF['Joy'], y=BlendDF['overall'], data=BlendDF, ax=axs[0], scatter = False, color = 'm')
s =sns.regplot(x=BlendDF['Sadness'], y=BlendDF['overall'], data=BlendDF, ax=axs[1], scatter = False, color = 'r')
a =sns.regplot(x=BlendDF['Anger'], y=BlendDF['overall'], data=BlendDF, ax=axs[2], scatter = False, color = 'g')
d = sns.regplot(x=BlendDF['Disgust'], y=BlendDF['overall'], data=BlendDF, ax=axs[3], scatter = False, color ='b')
f = sns.regplot(x=BlendDF['Fear'], y=BlendDF['overall'], data=BlendDF, ax=axs[4], scatter = False, color ='c')
j.set(ylim=(0, 5))
s.set(ylim=(0, 5))
a.set(ylim=(0, 5))
d.set(ylim=(0, 5))
f.set(ylim=(0, 5))
fig.suptitle ("Linear Relationship of the 5 Different Total Emotion Scores and Overall Ratings")
```

```python
plt.figure(figsize=(12,10))
fig, axs = plt.subplots(ncols=5)
j  = sns.regplot(x=BlendDF['Joy'], y=BlendDF['BinaryRating'], data=BlendDF, ax=axs[0], scatter = False, color = 'm')
s =sns.regplot(x=BlendDF['Sadness'], y=BlendDF['BinaryRating'], data=BlendDF, ax=axs[1], scatter = False, color = 'r')
a =sns.regplot(x=BlendDF['Anger'], y=BlendDF['BinaryRating'], data=BlendDF, ax=axs[2], scatter = False, color = 'g')
d = sns.regplot(x=BlendDF['Disgust'], y=BlendDF['BinaryRating'], data=BlendDF, ax=axs[3], scatter = False, color = 'b')
f = sns.regplot(x=BlendDF['Fear'], y=BlendDF['BinaryRating'], data=BlendDF, ax=axs[4], scatter = False, color = 'c')
j.set(ylim=(0, 1))
s.set(ylim=(0, 1))
a.set(ylim=(0, 1))
d.set(ylim=(0, 1))
f.set(ylim=(0, 1))
fig.suptitle ("Linear Relationship of the 5 different Total Emotions and Overall Ratings")
```

```python
plt.figure(figsize=(12,10))
fig, axs = plt.subplots(ncols=2)
j  = sns.regplot(x=BlendDF['VaderCompound'], y=BlendDF['overall'], data=BlendDF, ax=axs[0], scatter = False, color = 'c')
s =sns.regplot(x=BlendDF['VaderCompound'], y=BlendDF['BinaryRating'], data=BlendDF, ax=axs[1], scatter = False)

j.set(ylim=(0, 5))
s.set(ylim=(0, 1))
fig.suptitle ("Linear Relationship of VADER Score with Overall and Binary Ratings")
```

```python
plt.figure(figsize=(20,3))
from yellowbrick.features import rank1d, rank2d
X = BlendDF[['VaderCompound', 'Short', 'Verified', 'Long', 'IsImage',
       'WordCount', 'vote', 'LuxBea', 'VidGam', 'Joy', 'Anger', 'Sadness',
       'Fear', 'Disgust', 'overall', 'BinaryRating']]
_, axes = plt.subplots(ncols=2, figsize=(8,4))

rank1d(X, ax=axes[0], show=False)
rank2d(X, ax=axes[1], show=False)
plt.show()
plt.tight_layout()
```

```python
from yellowbrick.features import Rank1D

visualizer = Rank1D(algorithm='shapiro')
X = BlendDF[['VaderCompound', 'Short', 'Verified', 'Long', 'IsImage',
       'WordCount', 'vote', 'LuxBea', 'VidGam', 'Joy', 'Anger', 'Sadness',
       'Fear', 'Disgust', 'overall','BinaryRating']]
visualizer.fit(X)           
visualizer.transform(X)        
visualizer.show()              
```

```python
from yellowbrick.features import Rank2D

X = BlendDF[['VaderCompound', 'Short', 'Verified', 'Long', 'IsImage',
       'WordCount', 'vote', 'LuxBea', 'VidGam', 'Joy', 'Anger', 'Sadness',
       'Fear', 'Disgust', 'overall','BinaryRating']]

visualizer = Rank2D(algorithm='pearson')

visualizer.fit(X)           
visualizer.transform(X)        
visualizer.show()              
```
